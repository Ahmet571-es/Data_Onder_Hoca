### Birden çok şekil gösterimi : 

```python
x_min = -10
x_max = 10
xn = 21
x = np.linspace(x_min, x_max, xn)
x
```

```python
fig, ((axes1,axes2,axes3)) = plt.subplots(1,3, figsize=(11,4)) # 1 satır 3 sutunluk şekil oluştur. figsize=(yükseklik,genişlik)
axes1.plot(x,x)
axes2.plot(x, x**2-20)
axes3.plot(x, x**3/20)
plt.tight_layout(pad=5) # arasında ki mesafeyi ayarlar. pad aralarındaki mesafeyi açar.
plt.show()
```

```python
# 2 satır 2 sutunluk grafik çizelim :
fig, ((axes1,axes2),(axes3,axes4)) = plt.subplots(2,2,figsize=(10,6))
axes1.plot(x,x)
axes2.plot(x, x**2-20)
axes3.plot(x, x**3/20)
axes4.plot(x, x**4/45)
plt.tight_layout(pad=5, h_pad=6)
plt.show()
```

```python
fig, ((axes1,axes2),(axes3,axes4)) = plt.subplots(2,2,figsize=(10,6))
axes1.plot(x,x,"bs")
axes2.plot(x, x**2-20,"g^")
axes3.plot(x, x**3/20,"o")
axes4.plot(x, x**4/45,"+")
plt.tight_layout(pad=5, h_pad=6)
plt.show()
```

```python
import math
```

```python
def bivariate_normal(x,y,rho) : 
    v = (x**2 + y**2 - 2*rho*x*y) / (2.0 * (1.0 - rho**2))
    z = np.exp(-v) / ((2.0 * math.pi * (1.0 - rho**2))**0.5)
    return z
```

```python
# 3 boyutlu bir grafik elde etmek için x,y değeri verince z değeri üreten bir fonksiyon
```

```python
ax = plt.axes(projection="3d")
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y) # x,y arraylerini kullanarak X ve Y matrisi oluşturduk.
rho = 0.7
Z = bivariate_normal(X,Y,rho)

ax.plot_surface(X,Y,Z)
plt.show()
```

```python
ax = plt.axes(projection="3d")
ax.plot_surface(X,Y,Z, cmap=plt.cm.coolwarm)
plt.show()
```

```python
ax = plt.axes(projection="3d")
#ax.plot_surface(X,Y,Z, cmap=plt.cm.coolwarm)
ax.plot_surface(X,Y,Z, rstride=2, cstride=2, cmap=plt.cm.coolwarm)
plt.show()
```

```python
ax = plt.axes(projection="3d")
ax.plot_wireframe(X,Y,Z, rstride=5, cstride=5, color="green")
plt.show()
# rstride=5, cstride=5 noktaları daha iyi görmek istersek biraz daha arttırırız. Önce bir şeklin noktaları bulunur. 
# sonra bu noktalar üzerine 3boyut gösterimi yapılır.
```

```python
ax = plt.axes(projection="3d")
ax.contourf(X,Y,Z)
plt.show()
```
